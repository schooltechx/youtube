4. แบ่งข้อมูลเป็นชุดสำหรับการ Train

```python
# นำเข้า load_iris จาก sklearn
from sklearn.datasets import load_iris
# นำเข้า train_test_split จาก sklearn
from sklearn.model_selection import train_test_split
# แบ่งข้อมูลเป็นสองส่วนสำหรับ Train และ Test จะ 75%/25% อัตโนมัติ 
x_train, x_test, y_train,y_test = train_test_split(
    iris_dataset['data'],
    iris_dataset['target'],
    random_state=0
)
# ดูข้อมูลทั้งหมดมี 150 ชุด 75%=112, 25%=38
print('x_train shape:', x_train.shape)
print('x_test shape:', x_test.shape)
```
