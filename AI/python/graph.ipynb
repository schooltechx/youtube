```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('datasets/gdp.csv')
plt.xticks(df['year'])
plt.xlabel('Year')
plt.ylabel('GDP (Billion Bath)')
plt.plot(df['year'],df['gdp'])
plt.show()
plt.bar(df['year'],df['gdp'],color=['green','red','blue','yellow'],edgecolor='gray')
plt.show()
plt.title('Thailand GDP')
plt.pie(df['gdp'],labels=df['year'],autopct='%.2f%%',colors=['green','red','blue','yellow'] )
plt.show()
plt.scatter(df['year'],df['gdp'])
plt.xticks(df['year'])
plt.show()
```
